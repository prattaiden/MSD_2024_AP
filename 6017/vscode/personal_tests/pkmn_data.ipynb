```python
#add stats together to ger a rank of the pokemon 

pkmn['Total'] = pkmn['HP'] + pkmn['Attack'] + pkmn['Defense'] + pkmn['Sp. Atk'] + pkmn['Sp. Def'] + pkmn['Speed']

pkmn.sort_values('Total', ascending=False).head(10)
#display(pkmn)
```

```python
pkmn = pkmn.drop(columns = ['Total'])
#display(pkmn)

pkmn['Total Stats'] = pkmn.iloc[:, 4 : 10].sum(axis=1) #axis = 1 adds them horizontally, 0 is vertically 
cols = list(pkmn.columns.values)
pkmn = pkmn[cols[0:4] + [cols[-1]]+cols[4:12]]

pkmn.head(5)

```

```python
#advanced filtering of the data 
strong_grass = pkmn.loc[(pkmn['Type 1']== 'Grass') & (pkmn['Attack'] > 100)]
#drops the old index and makes a new one
strong_grass.reset_index(drop = True, inplace=True)

strong_grass

```

```python
#filter out names that contain mega !
# ~ removes them
no_mega = pkmn.loc[~pkmn['Name'].str.contains('Mega')]
no_mega


```

```python
import re #regular expressions 

#search for pokemon that have a type1 that is only fire or steel, ignore the capitals
pkmn.loc[pkmn['Type 1'].str.contains('fire|steel', flags=re.I,regex=True)]

```

```python
#pokmeon that start with some letters

# ^ specifies the start of the name * specifies 0 or more chars from a-z
pkmn.loc[pkmn['Name'].str.contains('^pi[a-z]*', flags = re.I, regex = True)]
```

```python
#conditional changes

#all fire types are legendary now
pkmn.loc[pkmn['Type 1']=='Fire', 'Legendary'] = True
pkmn
```
